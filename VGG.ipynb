모델 사용

```python
import tensorflow.keras.preprocessing.image as p
from tensorflow.keras.applications.vgg16 import preprocess_input, decode_predictions
```

```python
model_vgg16 = vgg16.VGG16()
```

```python
model_vgg16.save('vgg_model.h5')
```

```python
testimg = p.load_img('/content/test2.jpg',target_size=(224,224)) # 모델에 따라 target size 변경경

imgArr = p.img_to_array(testimg)
```

```python
pred = model_vgg16.predict(preprocess_input(imgArr.reshape(1,224,224,3)))

decode_predictions(pred)
```

전이 학습

```python
model_vgg16 = vgg16.VGG16(include_top=False,input_shape=(98,98,3)) # include_top=False: 분류기쪽만 추출
```

```python
model_vgg16.summary()
```

```python
dataGen = p.ImageDataGenerator(rescale=1./255)
trainGen = dataGen.flow_from_directory('/content/drive/MyDrive/data/myimg',target_size=(98,98))
```

```python
for layer in model_vgg16.layers:
  layer.trainable=False
```

```python
model = Sequential()
model.add(model_vgg16)
model.add(Flatten())
model.add(Dense(units=64,activation='relu'))
model.add(Dense(units=2,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics='acc')
model.fit(trainGen,epochs=10)
```
