# Step1. 기본 세팅

```python
from tensorflow.keras.datasets import boston_housing
```

```python
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()
```

```python
x_train.shape, x_test.shape
```

```python
y_train.shape, y_test.shape
```

# Step2. 학습 데이터의 전처리
* feature
  * standardization

```python
mean = x_train.mean(axis=0)

std = x_train.std(axis=0)
```

```python
x_train = (x_train - mean) / std

x_test = (x_test - mean) / std
```

# Step3. Model 구조 설계
*  input shape (13,)
* 1개의 연속적인 값을 예측하는 회귀 모델 구현

```python
from tensorflow.keras import models, layers
```

```python
model = models.Sequential()
```

```python
model.add(layers.Dense(units=64,activation='relu',input_shape=(13,))) # 초기 layer이므로 input_shape 지정
model.add(layers.Dense(units=32,activation='relu'))
model.add(layers.Dense(units=1)) # 집값 1개 이므로 perceptron은 1개, 활성 함수는 항등함수(default)
```

```python
model.summary()
```

```python
from tensorflow.keras.utils import plot_model
```

```python
plot_model(model=model, show_shapes=True)
```

# Step4. Model의 학습에 필요한 정보 세팅

```python
model.compile(loss='mse',optimizer='rmsprop',metrics=['mae'])
```

# Step5. Model에 학습데이터를 연결하여 학습

```python
history = model.fit(x=x_train, y=y_train, epochs=100, batch_size=32,validation_split=0.2)
```

# Step6. Test 데이터 셋을 이용한 모델의 성능 평가

```python
loss, mae = model.evaluate(x = x_test, y =  y_test)
```
