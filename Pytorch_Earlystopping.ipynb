```python
import torch
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.nn import Sequential,Linear,CrossEntropyLoss,Softmax, BCELoss
from torch.optim import Adam
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
```

```python
class MyData(Dataset):
  def __init__(self):
    self.x_data = [[10,20],[30,40],[33,43],[50,60],[70,80],[90,92]]
    self.y_data = [[100],[110],[115],[130],[135],[140]]

  def __len__(self):
    return len(self.x_data)
  
  def __getitem__(self,idx):
    x = torch.FloatTensor(self.x_data[idx])
    y = torch.FloatTensor(self.y_data[idx])
    return x,y
```

```python
dataset = MyData()
dataloader = DataLoader(dataset,batch_size=2,shuffle=True)
```

```python
model = torch.nn.Linear(2,1)
optimizer = torch.optim.Adam(model.parameters(),lr=0.01)
```

```python
for epoch in range(10):
  for batch_idx,samples in enumerate(dataloader):
    print(batch_idx)
    print(samples)
    print("~~~~~~~~~~~~~~~~")
    x_train,y_train = samples
    hx = model(x_train)
    cost = F.mse_loss(hx,y_train)
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    print(epoch,cost.item())
  print("==================")
```

```python
iris = load_iris()
```

```python
x_data = iris['data']
y_data = iris['target']

x = torch.FloatTensor(x_data)
y = torch.FloatTensor(np.eye(3)[y_data])
```

```python
model = Sequential()
model.add_module('nn1',Linear(4,3))
model.add_module('soft1',Softmax(dim=1))
loss_fn = CrossEntropyLoss()
optimizer = Adam(model.parameters(),lr=0.1)
```

```python
patience = 100
min_delta = 0.01
hist_cost = []
pcnt = 0

for epoch in range(2000):
  optimizer.zero_grad()
  hx = model(x)
  c = loss_fn(hx,y)
  c.backward()
  optimizer.step()
  hist_cost.append(c.item())
  print(epoch,c.item())

  if epoch > 0:

    if hist_cost[epoch-1] - hist_cost[epoch] > min_delta:
      
      pcnt = 0

    else:

      pcnt += 1

    if pcnt >= patience:

      print('early stop')

      break
```
